**The Normal Distribution has:**


1.   mean = median = mode
2.   symmetry about the center
3.   50% of values less than the mean and 50% greater than the mean

```python
import random
import plotly.express as px
import plotly.figure_factory as ff
count=[]
totalValue=[]

for i in range(0,1000):
  dice1= random.randint(1,6)
  dice2= random.randint(1,6)
  totalValue.append(dice1+dice2)
  count.append(i)

print(count)
print(totalValue)

graph= px.bar(x=totalValue, y=count)
graph.show()




```

```python
#  calculating Mean
mean= sum(  totalValue)/len(totalValue)


# Mode
import statistics
mean=statistics.mean(totalValue)
print(mean)
Mode= statistics.mode(totalValue)
print(Mode)

Median= statistics.median(totalValue)
print(Median)
```

```python
import plotly.figure_factory as ff
graph1= ff.create_distplot([totalValue],["Result"], show_hist=False)
graph1.show()
```

**Mean, median and mode are all
equal in a normal distribution and they
are at the peak!
This is a property common to all the
normal distributions!**


Normal Distribution has other interesting features too! Let's calculate the standard deviation (sd) for the data set we have.

```python
#  Calculating standrad Deviation using statistic module
import statistics
standard_Deviation= statistics.stdev(totalValue)
print(standard_Deviation)

```

Now, let's say we want to find what percentage of data lie between one standard deviation from the mean, i.e; what percentage of data lie between mean - sd and mean + sd ; how would you do that?

```python
# Finding 1 standard deviation start and end values, and 2 standard deviations start and end values

first_std_deviation_start, first_std_deviation_end= mean-standard_Deviation, mean+standard_Deviation
print(first_std_deviation_start, first_std_deviation_end)

second_std_deviation_start,second_std_deviation_end= mean-(2*standard_Deviation),mean+(2*standard_Deviation)
print(second_std_deviation_start,second_std_deviation_end)

Third_std_deviation_start,Third_std_deviation_end= mean-(3*standard_Deviation),mean+(3*standard_Deviation)
print(Third_std_deviation_start,Third_std_deviation_end)
```

```python
#  Drawing the graph to how much percentage of data is lies btw standard deviation
import plotly.graph_objects as go

# bell curve
graph2= ff.create_distplot([totalValue],["Result"],show_hist=False)
graph2.add_trace(go.Scatter(x=[first_std_deviation_start,first_std_deviation_start],y=[0,0.15], mode="lines", name="standard_deviation1"))
graph2.add_trace(go.Scatter(x=[first_std_deviation_end,first_std_deviation_end], y=[0,0.15], mode="lines",name="standard_deviation1"))
graph2.show()


```

```python
# printing Values
fianl_data_within_1st_std=[result for result in totalValue if result> first_std_deviation_start and result<first_std_deviation_end]
len(fianl_data_within_1st_std)
# #  total length final value *100/ total value 
# print("{} is a data lies inside the 1std".format(len(fianl_data_within_1st_std)*100.00/len(totalValue)))
```

```python
# normal distribution of Height and weight data
import pandas as pd
import plotly.figure_factory as ff
import plotly.graph_objects as go
import statistics
from statistics import mode

#  read the data
df= pd.read_csv("https://raw.githubusercontent.com/whitehatjr/Properties-of-normal-distribution/master/height-weight.csv")
print(df.head())

```

```python
#  get Height data
Height_data=df["Height(Inches)"].tolist()

# calculate mean, mode and median of Height
Height_mean=statistics.mean(Height_data)
Height_median=statistics.median(Height_data)
# Height_mode=statistics.mode(Height_data)

print("Mean,Meadian, and Mode of height is {},{} ".format(Height_mean,Height_median))
```

```python
#Standard deviation for height 
Height_std_deviation = statistics.stdev(Height_data)
print(Height_std_deviation)
```

```python
#1, 2 and 3 Standard Deviations for height
Height_first_std_deviation_start, Height_first_std_deviation_end = Height_mean-Height_std_deviation,Height_mean+Height_std_deviation
Height_second_std_deviation_start, Height_second_std_deviation_end = Height_mean-(2*Height_std_deviation), Height_mean+(2*Height_std_deviation)
Height_third_std_deviation_start,Height_third_std_deviation_end = Height_mean-(3*Height_std_deviation),Height_mean+(3*Height_std_deviation)


print(Height_first_std_deviation_start, Height_first_std_deviation_end)
print(Height_second_std_deviation_start, Height_second_std_deviation_end)
print(Height_second_std_deviation_start, Height_second_std_deviation_end)

```

```python
#Percentage of data within 1, 2 and 3 Standard Deviations for Height
Height_list_of_data_within_1_std_deviation = [result for result in Height_data if result > Height_first_std_deviation_start and result < Height_first_std_deviation_end]
Height_list_of_data_within_2_std_deviation = [result for result in Height_data if result > Height_second_std_deviation_start and result < Height_second_std_deviation_end]
Height_list_of_data_within_3_std_deviation = [result for result in Height_data if result > Height_third_std_deviation_start and result < Height_third_std_deviation_end]

# Printing data for height (Standard Deviation)
print("{}% of data for height lies within 1 standard deviation".format(len(Height_list_of_data_within_1_std_deviation)*100.0/len(Height_data)))
print("{}% of data for height lies within 2 standard deviations".format(len(Height_list_of_data_within_2_std_deviation)*100.0/len(Height_data)))
print("{}% of data for height lies within 3 standard deviations".format(len(Height_list_of_data_within_3_std_deviation)*100.0/len(Height_data)))

```

```python
#  get weight data
Weight_data=df["Weight(Pounds)"].tolist()
# print(Weight_data)


# calculate mean, mode and median of Height
Weight_mean=statistics.mean( Weight_data)
Weight_median=statistics.median( Weight_data)
Weight_mode=statistics.mode( Weight_data)
# Height_mode=statistics.mode(Height_data)

print("Mean,Meadian, and Mode of height is {},{},{} ".format (Weight_mean, Weight_median,Weight_mode))
```
